```python
import pandas as pd
import numpy as np
import folium
import geopy.distance

import ast
```

```python
# Load the Dataset
df = pd.read_csv('https://drive.google.com/uc?id=19JERBe0ndadkMPKDFZ0rTEZqdI50zDfk')
```

```python
df.head()
```

```python
# Merge Columns
df['Coordinate'] = '(' + df['lat'].astype(str) + ',' + df['lng'].astype(str) + ')'
df.head()
```

```python
# New York Empire State Building Lat Lng (center point)
NY_coord = (40.74855426303021, -73.98562148839889)
```

```python
def tuple_strings_to_floats(tup):
    return tuple(float(val) for val in ast.literal_eval(tup))

# Apply the conversion function to the DataFrame column
df['Coordinate'] = df['Coordinate'].apply(tuple_strings_to_floats)
```

```python
for sd in df.itertuples(index=True):
  print(len(sd))
  print(type(sd))
  d=geopy.distance.distance(NY_coord, sd.Coordinate)
  dff=geopy.distance.distance(NY_coord, sd.Coordinate).km
  print(d,dff)
  print(type(dff))
  break
```

```python
# Compute distance of each store from city center
distances_km = []

for row in df.itertuples(index = False):
  distances_km.append(
      geopy.distance.distance(NY_coord, row.Coordinate).km
  )
df['Distance'] = distances_km
df.head()
```

### Visualize data on a map

```python
# Empire State Bilding coordinates

m = folium.Map(location = NY_coord, zoom_start = 15)

# Place markers for the stores on the map

for i, row in df.iterrows():
  lat = df.at[i, 'lat']
  lng = df.at[i, 'lng']
  store = df.at[i, 'store']

  if store == 'McDonalds':
    color = 'red'
  else:
    color = 'green'

  folium.Marker(location=[lat,lng], popup=store, icon= folium.Icon(color=color)).add_to(m)

m
```

### Compute the closest and farthest stores from Empire State Building

```python
df_farthest = df.iloc[df.groupby('store')['Distance'].agg(pd.Series.idxmax)]
df_closest = df.iloc[df.groupby('store')['Distance'].agg(pd.Series.idxmin)]
df_new = pd.concat([df_farthest, df_closest])
```
